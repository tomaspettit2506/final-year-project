# AI Model Graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
pawn: 100,
  knight: 320,
  bishop: 330,
  rook: 500,
  queen: 900,
  king: 20000
"""

# Sample data
x = np.array([1, 2, 3, 4, 5, 6])  # Number of pieces
y = np.array([100, 320, 330, 500, 900, 20000])  # Corresponding values

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Chess Piece Values')
plt.xlabel('Number of Pieces')
plt.ylabel('Value')
plt.xticks(x, ['Pawn', 'Knight', 'Bishop', 'Rook', 'Queen', 'King'])
plt.grid()
plt.show()
```

```python
# AI thinking time is adjusted by difficulty
difficulty_levels = ['Easy', 'Medium', 'Hard', 'Expert', 'Master', 'Rocket']
thinking_times = [1, 3, 5, 7, 10, 15]  # in seconds

plt.figure(figsize=(10, 6))
plt.bar(difficulty_levels, thinking_times, color=['green', 'orange', 'red', 'purple', 'blue', 'black'])
plt.title('AI Thinking Time by Difficulty Level')
plt.xlabel('Difficulty Level')
plt.ylabel('Thinking Time (seconds)')
plt.grid(axis='y')
plt.show()
```
